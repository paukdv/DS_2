##### Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
top_languages = df['Язык.программирования'].value_counts().nlargest(3).index

df_top_languages = df[df['Язык.программирования'].isin(top_languages)]

plt.figure(figsize=(14, 8))
sns.countplot(data=df_top_languages, x='Город', hue='Язык.программирования',
              order=df_top_languages['Город'].value_counts().index)
plt.title('Розподіл мов програмування за регіонами України', fontsize=16)
plt.xlabel('Місто', fontsize=14)
plt.ylabel('Кількість спостережень', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Мова програмування', title_fontsize='12', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

```python
language_distribution = df['Язык.программирования'].value_counts(normalize=True)

filtered_languages = language_distribution[language_distribution >= 0.03]

plt.figure(figsize=(10, 10))
plt.pie(filtered_languages, labels=filtered_languages.index, autopct='%1.1f%%',
        startangle=90, counterclock=False, colors=sns.color_palette('pastel'))
plt.title('Частка мов програмування', fontsize=16)
plt.show()
```
